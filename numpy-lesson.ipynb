**Basic Array Operations**

```python
# One dimensional array
a = np.array([5,6,9])
print(a)

# print dimensions of an array .ndmim
print(a.ndim)
```

```python
# Two dimensional array
a = np.array([[1,2],[3,4],[5,6]])

print(a)
print(a.ndim)

# item size in bytes
print(a.itemsize)

#change data type
a = np.array([[1,2],[3,4],[5,6]], dtype=np.float64)
print(a.itemsize)
```

```python
# size

print(a.size)

print(a)

# shape  see information of dimensions rows - columns

print(a.shape)

# complex numbers

a = np.array([[1,2],[3,4],[5,6]], dtype=complex)

print(a)

a = np.array([[1,2],[3,4],[5,6]], dtype=int)

print(a)

print(a.itemsize)
```

```python
# zeros()   ones()

import numpy as np

a1 = np.zeros( (3,4) )
print(a1)

a2 = np.ones( (4,3) )
print(a2)
```

```python
# range of array

a1 = np.arange(1,5)
print(a1)

# range with steps

a2 = np.arange(1,5,2) #step = 2 so it gives 1 and 3
print(a2)

# linspace

a3 = np.linspace(1,10,10) # start number, stop number, how many numbers linear
print(a3)
```

**Mathematical functions**

```python
import numpy as np

a = np.array([[1,2],[3,4],[5,6]])
print("")

# minimum

print(a.min())
print("")

# maximum

print(a.max())
print("")

# summary of total and axis

print(a)
print(a.sum())
print("")
print(a.sum(axis=0))    # axis 0 vertical
print("")
print(a.sum(axis=1))    # axis 1 horizontal
print("")

# square root

print(np.sqrt(a))
print("")

# standard deviation

print(np.std(a))
print("")
```

**Slicing / Stacking & Indexing with boolean arrays**

```python
import numpy as np

# slice list
n = [6,7,8]
print(n[0:2])

print("")

# slice array
a = np.array([6,7,8])
print(a[0:2])
print("")

a = np.array([[6,7,8],[1,2,3],[9,3,2]])
print(a)
print("")
print(a[0,2]) # row 0 - collumn 2
print("")
print(a[0:2,2]) # row 0 - collumn 2  AND  row 1 - collumn 2  [0:2] -> 0 and 1
print("")
print(a[-1]) # last row
print("")
print(a[-1, 0:2]) # last row - collumn 0 AND last row - collumn 1
print("")
print(a[:,1:3]) # all rows but collumns ONLY 1 and 2
```

```python
import numpy as np

a = np.array([[6,7,8],[1,2,3],[9,3,2]])

for row in a:
  print(row)
print("")

# flat   print all as a cell

for cell in a.flat:
  print(cell)
print("")

```

```python
from numpy.core.multiarray import result_type
import numpy as np

# create 2 arrays

a = np.arange(6).reshape(3,2)
b = np.arange(6,12).reshape(3,2)

print(a)
print("")
print(b)
print("")

# vertical stack of 2 arrays
print(np.vstack((a,b)))
print("")

# horizontal stack of 2 arrays
print(np.hstack((a,b)))
print("")


c = np.arange(30).reshape(2,15)
print(c)
print("")

# split horizontal in 3 arrays

result = np.hsplit(c,3)
print(result)
print("")
print(result[0])
print("")
print(result[1])
print("")
print(result[2])
print("")

# split vertical in 3 arrays

c = np.arange(30).reshape(2,15)

result = np.vsplit(c,2)

print(result)
```

**Iterate numpy array**

```python
import numpy as np

a = np.arange(12).reshape(3,4)

print(a)
print("")


for x in np.nditer(a,order='c'):    # c order
  print(x)
print("")

for x in np.nditer(a,order='f'):    # fortran order
  print(x)
print("")

for x in np.nditer(a,order="f", flags=["external_loop"]):
  print(x)
print("")

print("=========================================")
for x in np.nditer(a,op_flags=["readwrite"]):    # make square of all numbers
  x[...]= x * x
print(a)
print("")

b = np.arange(3,15,4).reshape(3,1)

print(b)
print("")

for x,y in np.nditer([a,b]):
  print(x,y)
```
